```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('spotify-2023.csv', encoding='latin-1')
df
```

```python
df.describe()
```

```python
df = df.drop(574)
df[:575]
```

```python
df.filter(like='%').describe()
```

```python
df['streams'] = df['streams'].astype('int64')
df.dtypes
```

```python
condicao = df['streams'] > 1000000000
df_filtro = df[condicao].sort_values(by='streams', ascending=False).head(10)
df_filtro
```

BARPLOT DAS MÚSICAS MAIS OUVIDAS

- MÚSICA MENOS OUVIDA ESTÁ ACIMA DE 2 BILHÕES DE STREAMS;
- SOMENTE O TOP 1 E A TOP 2 ESTÁ ACIMA DE 3 BILHÕES;

```python
plt.subplots()
sns.barplot(df_filtro, x = 'track_name', y = 'streams')
plt.xticks(df_filtro['track_name'], rotation = 90)
```

BOXPLOT DA DANÇABILIDADE DAS 10 MÚSICAS MAIS OUVIDAS:

MÉDIA DE DANÇABILIDADE: 67%

 - AS 10 MÚSICAS MAIS OUVIDAS ESTÃO DENTRO DA MÉDIA DE DANÇABILIDADE;
 - A MÚSICA MAIS OUVIDA ESTÁ ABAIXO DA MÉDIA;
 - O TOP 10, TEM PELO MENOS UMA TAXA DE 50% DE DANÇABILIDADE;

```python
sns.boxplot(df_filtro['danceability_%'])
```

BARPLOT DO NÚMERO DE PLAYLISTS DO SPOTIFY QUE A MÚSICA SE ENCONTRA:

- A MÚSICA MAIS OUVIDA ESTÁ EM MAIS PLAYLIST'S
- A MÚSICA MENOS OUVIDA ESTÁ ENTRE AS TOP 3 EM PLAYLISTS;

```python
sns.barplot(df_filtro, x = 'track_name', y = 'in_spotify_playlists')
plt.xticks(df_filtro['track_name'], rotation = 90)
```

HISTPLOT BPM

MÉDIA BPM: 122

- TEMOS 3 FAIXAS DE BPM;
- A MÚSICA MAIS OUVIDA E A MENOS OUVIDA ESTÁ NA FAIXA MAIS ALTA DE BPM;
- O BPM NÃO INTERFERE DIRETAMENTE NO SUCESSO DA MÚSICA;

```python
sns.histplot(df_filtro, x = 'track_name', y = 'bpm')
plt.xticks(df_filtro['track_name'], rotation = 90)
```

LINEPLOT DAS MÚSICAS:

- A MÚSICA TER SIDO LANÇADA A MAIS TEMPO NÃO GARANTE MAIS STREAMS;
- A MÚSICA MENOS OUVIDA É A MAIS ANTIGA;

```python
sns.lineplot(df_filtro, x = 'released_year', y = 'streams')
```

```python
sns.barplot(df_filtro, x = 'streams', y= 'track_name', hue='in_spotify_playlists')
```
