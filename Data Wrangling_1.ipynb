```python
import pandas as pd
```

### Importando os DataFrames

```python
df1 = pd.read_csv('w_data.csv', sep=';') # importando os dados já separando as colunas
df1 #visualização
```

```python
df2 = pd.read_csv('p_data.csv',sep=';')  # importando os dados já separando as colunas
df2
```

```python
df_merge = pd.merge(df2,df1, on= 'ID') #Fundindo os dois DF's pelo ID
df_merge.head(10) 
```

```python
df_merge['ID'].value_counts().sort_values #conferindo a existências de ID's repetidos
```

```python
df_merge['Target'].unique() # conferindo os valores da coluna Target
```

```python
subs = {' <=50K.':' <=50K', ' >50K.':' >50K'} #dicionário para substituir os termos da coluna Target
```

```python
df_merge = df_merge.replace(subs) #Substituindo termos
```

```python
df_merge['Target'].unique() # conferindo se os valores foram substituidos na coluna Target
```

```python
filtro = (df_merge['Target'] == ' >50K') #Filtro para salários acima de 50K
```

Gênero

```python
sexo = df_merge[filtro]['Sex'].value_counts() #Filtro para verificar valores da coluna Sex acima de 50K
```

```python
df_sexo = pd.DataFrame(sexo) #Gerando um DF a partir dos dados filtrados e contabilizados
df_sexo
```

```python
porcentagem_sex = (df_sexo['count'] / df_sexo['count'].sum() * 100).round() #calculo de porcentagem
porcentagem_sex
```

```python
df_sexo['%'] = porcentagem_sex #incluindo a coluna porcentagem
df_sexo
```

Raça

```python
Race = df_merge[filtro]['Race'].value_counts() #Filtro para verificar valores da coluna Race acima de 50K
```

```python
df_race = pd.DataFrame(Race) #Gerando um DF a partir dos dados filtrados e contabilizados
df_race
```

```python
porcentagem_race = (df_race['count'] / df_race['count'].sum() * 100).round(2) #calculo de porcentagem
porcentagem_race
```

```python
df_race['%'] = porcentagem_race #incluindo a coluna porcentagem
df_race
```
