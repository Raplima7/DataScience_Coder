```python
import numpy as np ##importando numpy
import pandas as pd ##importando pandas
```

```python
df = pd.read_csv("Summer-Olympic-medals-1976-to-2008.csv",encoding='latin-1') ##Lendo arquivo original .csv
df
```

```python
df_olimpycs = df[['Year', 'City', 'Sport', 'Discipline', 'Event', 'Athlete', 'Gender',
      'Country_Code', 'Country', 'Event_gender', 'Medal']] ##Substituição de ordem da coluna 'City' com a coluna 'Year', para a coluna Year ser a primeira
df_olimpycs
```

```python
indice = list(df_olimpycs.Year) ##criando indice com a coluna 'Year'
colunas = df_olimpycs.columns[1:] ## determinando as colunas do novo DF excluindo a coluna 'Year'
valores = df_olimpycs.values[:,1:] ## determinando os valores das colunas
print(indice) #validando indice
colunas #validando colunas
```

```python

df_olimpycs2 = pd.DataFrame(valores,index=indice,columns=colunas) #criação do novo DF com o ano como indice
df_olimpycs2
```
