```python
df_data = pd.DatetimeIndex(df['DateofHire']) #transformando a coluna DateofHire em data
```

```python
df['Ano ContrataÃ§Ã£o'] = df_data.year #pegando somente o ano da data
df.head()
```

```python
df.index = df['Ano ContrataÃ§Ã£o'].sort_values() #definindo o index em ordem crescente
df
```
