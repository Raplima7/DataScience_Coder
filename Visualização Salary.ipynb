```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('HRDataset_v14.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df["Sex"].value_counts()
```

```python
mpl.style.use("ggplot")
```

```python
sns.kdeplot(data = df, x = "Salary", hue = "Sex")
```

```python
fig, ax = plt.subplots(figsize=(12,10))
ax.hist(df["Salary"])
```

```python
fig
ax.set_title("Sálarios mais populares")
ax.set_xlabel("Valor do Sálario em R$")
ax.set_ylabel("nº de funcionários")
fig
```

```python
ax1 = sns.catplot(data = df, kind='violin', y = df["Salary"], x = df["Sex"], split = True)
ax1.set(ylabel = 'Valor do Salário em R$', xlabel = "Gênero do funcionário", title = 'Faixa salarial por gênero')
ax1
```

```python
df_data = pd.DatetimeIndex(df['DateofHire']) #transformando a coluna DateofHire em data
```

```python
df['Ano Contratação'] = df_data.year #pegando somente o ano da data
df.head()
```

```python
df.index = df['Ano Contratação'].sort_values() #definindo o index em ordem crescente
df
```

```python
y = df.index.value_counts().sort_index() #visualizando valores para gerar gráfico
y
```

```python
fig1, ax1 = plt.subplots(figsize = (12,10)) #definindo tamanho da imagem
```

```python
ax1.plot(y) #plotando variável y
fig1 #exibindo série temporal baseado em quantidade de contratações por ano
```

```python
ax1.set_title('Nº Contratações x Ano') #definindo título do gráfico
ax1.set_xlabel('Ano') # definindo rótulo eixo X
ax1.set_ylabel('Nº Contratações') #definindo rótulo eixo Y
fig1
```
